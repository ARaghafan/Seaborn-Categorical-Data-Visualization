# Seaborn Categorical Data Visualization

## 1. Introduction

**What is this topic about?**  
Categorical data visualization is about representing data where the values are not numbers but categories, like gender, product type, or city. Seaborn, a Python visualization library built on top of Matplotlib, makes it very easy to create attractive and informative plots for such data.

**Short overview in my own words:**  
Instead of dealing with numbers like ‚Äúheights in centimeters,‚Äù categorical visualization focuses on groups like ‚Äúmale vs female‚Äù or ‚Äúapples vs oranges.‚Äù Seaborn provides special plots‚Äîlike bar plots, count plots, box plots, and violin plots‚Äîthat help us understand differences between these categories.

**Real-life examples:**
- Comparing sales of different product types in a store (electronics vs clothing vs food).  
- Visualizing survey responses like ‚Äúsatisfied, neutral, dissatisfied.‚Äù  
- Comparing average salaries across different professions.  

---

## 2. Theoretical Background

**Definition of categorical data:**  
Categorical data represents information that can be divided into groups or categories. Unlike numerical data, it doesn‚Äôt make sense to calculate things like averages directly (e.g., you can‚Äôt average ‚Äúmale‚Äù and ‚Äúfemale‚Äù).

**Why visualize categorical data?**  
Because it helps us quickly see patterns, trends, and differences between groups. For example, a bar plot can show which category is more frequent, while a box plot can show how numerical values vary across categories.

**Key Seaborn functions for categorical plots:**
- `sns.barplot()` ‚Üí shows average values per category.  
- `sns.countplot()` ‚Üí shows counts of each category.  
- `sns.boxplot()` ‚Üí shows distribution (median, quartiles, outliers).  
- `sns.violinplot()` ‚Üí similar to boxplot but adds density information.  
- `sns.stripplot()` and `sns.swarmplot()` ‚Üí show individual data points within categories.  

**Mini-theory of bar plots (example):**
- X-axis ‚Üí categories (like "male" and "female").  
- Y-axis ‚Üí numerical values (like average income).  
- Bars show the mean value for each category, with optional error bars (confidence intervals).  

**About the library (Seaborn):**  
Seaborn is built on top of Matplotlib but is designed for statistical graphics. It automatically calculates things like averages, error bars, and density estimates for categorical data. This makes it easier and faster to visualize data without writing a lot of code.

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

# Load the Penguins Dataset¬∂

````python
penguins = sns.load_dataset("penguins")
print("\n" + "="*60)
print(penguins.info())
print("="*60)
print(penguins.duplicated())
print("="*60)
print(penguins[penguins.isna().all(axis=1)])
print("="*60)
penguins.isna().sum()

````

# üß© 1. CountPlot ‚Äî Frequency of Categories
### Shows category counts and gender distribution ‚Äî simple but impactful
show how many penguins belong to each species.

````python
sns.set(style="whitegrid", palette="muted")
plt.figure(figsize=(8,5))
sns.countplot(x="species", data=penguins, hue="island")
plt.title("Number of Penguins by Species and Sex")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
````

# üéØ 2. BoxPlot ‚Äî Compare numerical distributions by category
### How a numerical variable varies across categorical groups.
### ‚Üí Gentoo penguins tend to have longer flippers.

````python
plt.figure(figsize=(8,6))
sns.boxplot(x="species", y="flipper_length_mm", hue="sex", data=penguins, palette="Set2")
plt.title("Flipper Length by Penguin Species and Sex")
plt.xlabel("Species")
plt.ylabel("Flipper Length (mm)")
plt.show()
````

# üé® 3. ViolinPlot ‚Äî Visualize shape of distribution
### Combines boxplot + density curve = visually rich & easy to explain.

````python
plt.figure(figsize=(8,6))
sns.violinplot(x="species", y="bill_depth_mm", hue="sex", data=penguins, split=True, palette="coolwarm")
plt.title("Body Mass Distribution by Penguin Species and Sex")
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")
plt.show()
````

# üí† 4. SwarmPlot ‚Äî Show individual observations
### Each dot = a penguin ‚Üí great to show real variability.

````python
plt.figure(figsize=(8,6))
sns.swarmplot(x="species", y="bill_length_mm", hue="sex", data=penguins, dodge=True)
plt.title("Bill Length by Species (individual observations)")
plt.xlabel("Species")
plt.ylabel("Bill Length (mm)")
plt.show()
````

# üß† 5. CatPlot ‚Äî Combine everything easily
### Multiple subplots in one command

### Compact, clear, and perfect for storytelling

````python
sns.catplot(
    x="species",
    y="bill_depth_mm",
    hue="sex",
    col="island",
    kind="box",
    data=penguins,
    height=5,
    aspect=0.8
)
plt.subplots_adjust(top=0.8)
plt.suptitle("Bill Depth by Species, Sex, and Island")
plt.show()
````

#  üêß 6. Pairing Categorical and Numerical
### ‚ú® Great for your presentation‚Äôs finale ‚Äî colorful, multi-variable overview.

````python
sns.pairplot(penguins, hue="species", diag_kind="kde", palette="husl")
plt.suptitle("Pairwise Relationships between Penguin Features", y=1.02)
plt.show()
````

# üß© 7. Barplot ‚Äî Show average values by category
### By default, sns.barplot() shows the mean of a numeric variable per category.
### Great to explain ‚Äúcentral tendency‚Äù ‚Äî i.e., which species are heavier on average.

````python
plt.figure(figsize=(8,6))
sns.barplot(x="species", y="body_mass_g", hue="sex", data=penguins, palette="viridis")
plt.title("Average Body Mass by Species and Sex")
plt.xlabel("Species")
plt.ylabel("Mean Body Mass (g)")
plt.show()
````

# üí¨ 8. PointPlot ‚Äî Trend or comparison across categories
### pointplot() adds confidence intervals and shows trends clearly.
### Why it‚Äôs impressive: It looks professional ‚Äî great for showing patterns or differences between groups.

````python
plt.figure(figsize=(8,6))
sns.pointplot(x="island", y="flipper_length_mm", hue="sex", data=penguins, palette="coolwarm", dodge=True)
plt.title("Average Flipper Length across Islands by Sex")
plt.xlabel("Island")
plt.ylabel("Mean Flipper Length (mm)")
plt.show()
````

# üìà 9. StripPlot ‚Äî Simpler version of swarmplot (faster)
### It‚Äôs like swarmplot but without collision avoidance ‚Äî quicker for big datasets.
### üéØ Use this when: You want to show individual data points but keep it simple

````python
plt.figure(figsize=(8,6))
sns.stripplot(x="species", y="body_mass_g", hue="sex", data=penguins, jitter=True, alpha=0.7, palette="Set1")
plt.title("Body Mass by Species and Sex (Strip Plot)")
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")
plt.show()
````

# üßÆ 10. BoxenPlot ‚Äî Advanced alternative to boxplot

### Boxen plots show distributional shape for large datasets ‚Äî more quantiles than boxplot.
### üí° Why use it: It‚Äôs visually richer and rarely shown ‚Äî will impress your classmates!

````python
plt.figure(figsize=(8,6))
sns.boxenplot(x="species", y="body_mass_g", hue="sex", data=penguins, palette="cubehelix")
plt.title("Body Mass Distribution by Species and Sex (Boxen Plot)")
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")
plt.show()
````

# üé® 11. CatPlot (violin type) ‚Äî quick grid visualization

### You can use sns.catplot() with different kind arguments (box, violin, bar, etc.):
### üéØ This is perfect for multi-category comparisons.

````python
sns.catplot(
    x="species",
    y="bill_length_mm",
    hue="sex",
    col="island",
    kind="violin",
    data=penguins,
    height=5,
    aspect=0.8,
    palette="muted"
)
plt.subplots_adjust(top=0.8)
plt.suptitle("Bill Length by Species, Sex, and Island (Violin CatPlot)")
plt.show()
````

# üí† 12. Countplot with multiple categories

### If you want a clean frequency plot (like a histogram for categorical data):
### üí° Why it‚Äôs strong: Everyone immediately understands counts and group differences.

````python
plt.figure(figsize=(8,6))
sns.countplot(x="island", hue="species", data=penguins, palette="pastel")
plt.title("Penguin Species Distribution across Islands")
plt.xlabel("Island")
plt.ylabel("Count")
plt.show()
````
